#Cálculo de Média

```python
#entrada
num1 = float(input('Digite o valor do numero 1: '))
num2 = float(input('Digite o valor do numero 2: '))
```

```python
#processamento
media = (num1+num2)/2
```

```python
#saida
print(f"A média é: {media}")
```

#Conversão de Moeda

```python
cotacaomoeda =  float(input('Digite o valor da cotacao do dolar: '))
valorreais =  float(input('Digite o valor: '))
```

```python
resultado = valorreais/cotacaomoeda
```

```python
print(f"O valor do dinheiro em dolar é: {resultado}")
```

#Cálculo de Comissão

```python
valor =   float(input('Digite o valor do item: '))
quantidade =   float(input('Digite a quantidades de itens vendidos: '))
```

```python
venda = valor*quantidade
comissao = (venda*5)/100
```

```python
print(f"O valor da venda foi: {venda}")
print(f"O valor da comissao foi de porcentagem: {comissao}")
```
